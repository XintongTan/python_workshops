```python
# importing numpy and pandas
import numpy as np
import pandas as pd

file_path = './data/IMDB Dataset.csv'

# reading the csv file into a pandas dataframe
df = pd.read_csv(file_path)

df.tail(10)
```

```python
df_test = df.sample(10000)

df_test
```

```python
df_new = pd.merge(df, df_test, how='outer', indicator=True)
print(df_new)
```

```python
df_exc = df_new.loc[df_new._merge == 'left_only', ['review','sentiment']]

df_exc
```

```python
df_training = df_exc.sample(30000)

df_training
```

```python
# reset the index
df_training.reset_index(drop=True, inplace=True)

df_training
```

```python
# reset the index
df_test.reset_index(drop=True, inplace=True)

df_test
```

```python
df_training.to_csv('./data/IMDB_training.csv')
```

```python
df_test.to_csv('./data/IMDB_test.csv')
```

```python
# what's the ratio between positive and negative reviews?
neg_count = len(df_training[df_training['sentiment']=='negative'])
print('negative reviews:', neg_count)

pos_count = len(df_training[df_training['sentiment']=='positive'])
print('positive reviews:', pos_count)
```
